```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('default')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data_arr = np.random.uniform(-1,1,100)
matrix = data_arr.reshape((10,10))
#matrix

plt.figure(figsize=(10,10))

heatmap = sns.heatmap(matrix, cmap = "Greens", vmin = -1, annot = True, vmax = 1)

plt.show()
```

```python
nrow = 1000
ncol = 3

data_arr_3000 = np.random.uniform(-1,1,3000)
matrix_3000 = data_arr_3000.reshape((1000,3))
matrix_3000

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix_3000, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)


grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap="Blues_d")

plt.show()
```

```python
nrow = 1000
ncol = 3

mu, sigma = 0, 0.1 # mean and standard deviation
data_arr_nor = np.random.normal(mu, sigma, 3000)
matrix_nor = data_arr_nor.reshape((1000,3))
matrix_nor

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix_nor, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap="Blues_d")

plt.show()
```
