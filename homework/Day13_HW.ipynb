```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
os.chdir(r"C:\Users\Wendy\Documents\GitHub\2nd-ML100Days\data")
```

### 之前做過的處理

```python
# 讀取資料檔
app_train = pd.read_csv("application_train.csv")
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 12 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
app_train['YEARS_BIRTH'] = app_train['DAYS_BIRTH'] / 365
app_train.head()
```

```python
app_train["equal_width_Y_age"] = pd.cut(app_train["YEARS_BIRTH"], [20,30,40,50,60,70])
app_train.equal_width_Y_age.value_counts()
```

```python
grouped_df = app_train.groupby('equal_width_Y_age')['TARGET']
```

```python
grouped_rdf = grouped_df.mean().reset_index()
grouped_rdf
```

```python
# width of the bars
barWidth = 0.5

# Choose the height of the blue bars
bars1 = grouped_rdf.TARGET
 
# The x position of bars
r1 = np.arange(len(bars1))
 
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=7)
 
# general layout
plt.xticks(r1, grouped_rdf.equal_width_Y_age)
plt.ylabel('TARGET')
plt.xlabel('Age') 
# Show graphic
plt.show()

# 年紀越大越可能不還錢
```
